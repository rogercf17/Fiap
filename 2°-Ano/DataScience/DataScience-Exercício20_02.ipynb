```python
# Exerc√≠cio 1 - Liste todos os her√≥is da editora Marvel.
df[df['Editora'] == 'Marvel']
```

```python
# Exerc√≠cio 2 - Qual √© o her√≥i mais antigo da lista?
data_mais_antiga = df['Ano_Estreia'].min()
df[df['Ano_Estreia'] == data_mais_antiga]
```

```python
# Exerc√≠cio 3 - Qual √© a m√©dia de for√ßa dos her√≥is da DC?
media_de_forca = df['For√ßa'].mean()
print(f'M√©dia de for√ßa de todos os h√©rois √©: {media_de_forca}')
```

```python
# Exerc√≠cio 4 - Qual editora tem a maior m√©dia de intelig√™ncia...
inteligencia = df.groupby('Editora')['Intelig√™ncia'].sum()
if inteligencia.DC > inteligencia.Marvel:
    print('DC √© a editora com mais intelig√™ncia entre os herois')
else:
     print('Marvel √© a editora com mais intelig√™ncia entre os herois')
```

```python
# Exrc√≠cio 5 - Qual √© o her√≥i mais veloz?
maior_velocidade = df['Velocidade'].max()
df[df['Velocidade'] == maior_velocidade]
```

```python
# Exerc√≠cio 6 - Quantos her√≥is estrearam antes de 1960?
herois_velhos = df[df['Ano_Estreia'] < 1960]

herois_velhos
```

```python
print(f'{herois_velhos.shape[0]} estrearam antes de 1960.')
```

```python
# Exerc√≠cio 7 - Qual √© a soma total da for√ßa dos her√≥is da Marvel
herois_marvel = df[df['Editora'] == 'Marvel']
forca_total = herois_marvel['For√ßa'].sum()
print(f'A soma das for√ßas dos h√©rois da marvel √© igual a {forca_total}')
```

```python
# Exerc√≠cio 8 - Liste os her√≥is em ordem decrescente de velocidade
df.sort_values(by='Velocidade', ascending=False)
```
