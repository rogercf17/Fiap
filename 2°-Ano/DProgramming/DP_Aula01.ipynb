## Série de Fibonacci

```python
def fibonacci(n):
  if n < 2:
    return n
  else:
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(5)
```

```python
def fatorial(n):
  fatorial_n = math.factorial(n)
  return fatorial_n
```

```python
resultado = fatorial(167)
resultado
```

```python
def soma(a, n):
    if n == 0:
        return a[0]
    return a[n] + soma(a, n - 1)
```

```python
soma([3,4,7,9,69], 3)
```
