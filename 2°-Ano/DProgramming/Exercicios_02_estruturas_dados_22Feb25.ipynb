## Exercício 1 - Média de cada dado

```python
def calcular_media(df, dado):
  valores = []
  for i in range(len(df)):
    if dado in df[i]:
        valores.append(df[i][dado])
  return int(sum(valores) / len(valores))
```

```python
calcular_media(data, 'bus')
```

```python
calcular_media(data, 'metro')
```

```python
calcular_media(data, 'taxi')
```

```python
calcular_media(data, 'bicycle')
```
